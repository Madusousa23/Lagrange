```python
#lagrange
print("analisando entradas...")
vetx=[-1,0,2]
vety=[4,1,-1]
x=-1
j=0
i=0
q=0
n=len(vetx)
for i in range (0,n):
  p=1
  for j in range(0,n):
    if j != i:
      p = p *(x - vetx[j])/(vetx[i]-vetx[j])
  q = q + p * vety[i]
print("resultado",q)
```

```python
#newton-em progresso
print("analisando entradas...")
print("")
vetx=[0,1,3,4]
vety=[1,-1,1,2]

n=len(vetx)

x0=0
termo=1
j=0

f=[[0 for j in range(n)]for i in range(n)]

for i in range (n):
  f[i][0] = vety[i]

for i in range (1,n):
  for j in range(i,n):
   f[j][i] = (f[j][i-1] - f[j-1][i-1]) / (vetx[j] - vetx[j-i])
print("Tabela de diferenças divididas: ")

for i in range (n):
  for j in range (i+1):
    print(round(f[i][j],3)) #acessar elementos da tabela de diferenças divididas| round é para arredondar 3 casas decimais



p=f[0][0]
for i in range(1,n):
  termo*=(x0 - vetx[i-1])
  p = p + f[i][i]*termo
print("")
print("Valor do polinômio em X0= ",x0,"é",p)
```
